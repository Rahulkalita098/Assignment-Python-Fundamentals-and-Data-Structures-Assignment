## Question 1: What is mutability in Python?

Mutability refers to whether an object can be changed after it is created.
Mutable objects can be modified in place, while immutable objects cannot.

Examples:
- Mutable: list, dictionary, set
- Immutable: tuple, string, integer

Example:
my_list = [1, 2, 3]
my_list[0] = 10  # Allowed because lists are mutable

## Question 3: What are logical operators in Python?

Logical operators are used to combine conditional statements.

1. and – Returns True if both conditions are True.
2. or – Returns True if at least one condition is True.
3. not – Reverses the result.

Example:
a = True
b = False

print(a and b)
print(a or b)
print(not a)

## Question 5: Describe how dictionaries store data.

A dictionary stores data as key-value pairs. Each key is unique and is used to access its corresponding value efficiently. Dictionaries are mutable and unordered (insertion order is preserved in modern Python versions).

Example:
student = {
    "name": "Rahul",
    "age": 22,
    "city": "Guwahati"
}

## Question 7: What is the difference between for and while loops?

For Loop:
- Used when the number of iterations is known.
- Iterates over a sequence.

While Loop:
- Used when the number of iterations is unknown.
- Continues until a condition becomes False.

Example:
for i in range(5):
    print(i)

count = 0
while count < 5:
    print(count)
    count += 1


## Question 8: Write a Python program to calculate the factorial of a number provided by the user using a while loop.

num = int(input("Enter a number: "))

factorial = 1
i = 1

while i <= num:
    factorial *= i
    i += 1

print("Factorial:", factorial)


## Question 9: Why might you use a set instead of a list in Python?

A set is useful when:
- Duplicate values are not allowed.
- Faster membership testing is needed.
- Mathematical set operations are required.

Example:
numbers = {1, 2, 3, 3, 4}
print(numbers)

Output:
{1, 2, 3, 4}
